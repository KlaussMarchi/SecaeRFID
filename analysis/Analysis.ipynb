```python
import requests
import pandas as pd
```

```python
url = r'https://xi-secae-default-rtdb.firebaseio.com/.json'
```

```python
response = requests.get(url, timeout=5)
response
```

```python
dataJson = response.json()
dataJson
```

```python
for activity, data in dataJson.items():
    print(activity + ':')
    df = []

    for info in data.values():
        df.append(info)

    df = pd.DataFrame(df)
    df.to_csv(f'target/{activity}.csv', index=None)
    display(df)
    print()
```
