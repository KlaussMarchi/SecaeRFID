```python
import pandas as pd
from pyperclip import copy
import numpy as np
```

```python
data = []

with open('data.txt', 'r', encoding='utf-8') as file:
    raw = file.readlines()

for line in raw:
    if len(line.strip()) == 0:
        continue

    if 'NOME' in line.upper():
        continue
    
    data.append(line.strip())

print(len(data))
data
```

```python
df = []

for line in data:
    row = line.split('\t')
    target = []

    for i, val in enumerate(row):
        val = val.strip()

        if len(val) < 5:
            continue

        target.append(val)
    
    if len(target) == 3:
        df.append(target)

df = pd.DataFrame(df, columns=['id', 'name', 'email'])
df
```

```python
df = df.sample(frac=1)#[:200]
df.head()
```

```python
step = 80
x = 0

stepList = list(np.arange(0, len(df), step)) + [len(df)]
stepList
```

```python
start = stepList[x]
end   = stepList[x+1]
x += 1

print('start:', start, '| end:', end, '| x:', x)
print()
text = ''

for email in df.email.values[start:end]:
    text += email + '\n'

copy(text)
print(text)
```
