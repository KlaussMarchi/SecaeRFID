```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_rows', None)
```

```python
database = {}

for file in os.listdir('activities'):
    df = pd.read_csv(f'activities/{file}')
    fileName = file.split('.')[0]
    
    for column in df.columns:
        df[column] = df[column].astype(str)
    
    if 'mat' in df.columns:
        df = df.drop_duplicates(subset=['mat'], keep='last').reset_index(drop=True)
        df = df.sort_values(by='name')
    
    database[fileName] = df

database['activities']
```

```python
for name, table in database.items():
    print(f'{name}: {len(table)} rows')
```

# CORRIGINDO ABERTURA
- As ultimas 60 pessoas da abertura são na verdade de outras palestras

```python
lastRows = database['abertura'].tail(60)
lastRows.head()
```

```python
database['simulador_rov'] = pd.concat([database['simulador_rov'], lastRows])
database['rov_tecnologia'] = pd.concat([database['rov_tecnologia'], lastRows])
database['candidato_desejado'] = pd.concat([database['candidato_desejado'], lastRows])
database['pesquisa_hidricos'] = pd.concat([database['pesquisa_hidricos'], lastRows])
database['eletronica_basica_i'] = pd.concat([database['eletronica_basica_i'], lastRows])
```

# JUNTANDO ABERTURA

```python
tables = [
    database['abertura'], 
    database['ieee'],
    database['maker'],
    database['atletica'],
    database['empresa_junior'],
]

remanecents = pd.concat(tables)
remanecents = remanecents.drop_duplicates(subset=['mat'], keep='last')

def setRemanecents(key):
    global database, remanecents
    df_copy = remanecents.copy() 
    df_copy['activitie'] = key   
    database[key] = df_copy   


setRemanecents('abertura')
setRemanecents('ieee')
setRemanecents('maker')
setRemanecents('atletica')
setRemanecents('empresa_junior')
len(remanecents)

database['abertura'].head()
```

# JUNTANDO ATIVIDADES

```python
df = pd.DataFrame()

for name, table in database.items():
    if name == 'users' or name == 'activities':
        continue
    
    table['activitie'] = name
    df = pd.concat([df, table])

df = df.drop_duplicates().reset_index(drop=True)
database['all'] = df
df.head()
```

```python
database['all'].activitie.unique()
```

```python
def smartSort(df, key):
    from unidecode import unidecode
    df['normalized'] = df[key].apply(lambda row: unidecode(row.lower()))
    df = df.sort_values(by='normalized').drop(columns=['normalized'])
    df = df.reset_index(drop=True)
    return df
```

```python
registration = database['users'].mat.unique()
print(len(registration))
users = []

for reg in registration:
    target = database['users'].loc[database['users'].mat == reg].iloc[-1]
    name   = target['name']
    mat    = target['mat'] 
    target = (df.mat == mat).sum()

    if target == 0:
        target = 1

    target = {'name': name, 'mat': mat, 'total': target}
    users.append(target)

users = pd.DataFrame(users)
users = smartSort(users, 'name')
database['presenca'] = users 
users.head(20)
```
