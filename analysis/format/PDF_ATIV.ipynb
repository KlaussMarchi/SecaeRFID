```python
for table in data:
    key, label, date = table['value'], table['label'], table['time']
    title = label + ' - ' + date

    latexCode += r"""
        \begin{center}
            \begin{minipage}{.85\linewidth}
            \centering
            \section{\LARGE $title!}
        \end{minipage}
        \end{center}

        \normalsize

        {\renewcommand{\arraystretch}{2}
        \begin{longtable}{|c|c|}
            \hline
                \multirow{1}{7cm}{\centering Nome} &
                \multirow{1}{6cm}{\centering Matrícula}\\
            \hline
    """.replace('$title!', title)

    target = df.loc[df.activitie == key]

    for i, row in target.iterrows():
        name = row['name']
        mat  = row['mat']
        
        latexCode += r"""
            \multirow{1}{7cm}{\centering $name!} &
                \multirow{1}{6cm}{\centering \large $mat!} \\
            \hline
        """.replace('$name!', name).replace('$mat!', mat)

    latexCode += r'\end{longtable}\newpage' + '\n'
```
