```python
import pandas as pd
from pyperclip import copy
```

```python
activitie = 'Palestra Comunicação Satélite'
```

```python
df = []

with open('inscritos.txt', 'r', encoding='utf-8') as file:
    data = file.readlines()

for line in data:
    row = line.split('\t')
    target = []

    for i, val in enumerate(row):
        val = val.strip()

        if len(val) < 5:
            continue

        target.append(val)
    
    if len(target) == 3:
        df.append(target)

df = pd.DataFrame(df, columns=['name', 'email', 'status'])
print(len(df))
df
```

```python
df = df.loc[~df.status.str.lower().str.contains('cancel')].reset_index(drop=True)
print(len(df))
df
```
