```python
import pandas as pd
from pyperclip import copy
```

```python
name = 'OneSubsea Imboassica'
```

```python
df = []

with open('inscritos.txt', 'r', encoding='utf-8') as file:
    data = file.readlines()

for line in data:
    row = line.split('\t')
    target = []

    for i, val in enumerate(row):
        val = val.strip()

        if len(val) < 3:
            continue

        target.append(val)
        df.append(target)

df = pd.DataFrame(df, columns=['name']).sort_values(by='name').reset_index(drop=True)
df['name'] = df.name.str.replace('-', ')').str.replace(',',')').str.split(')').str[1].str.strip()
df
```

```python
for i, row in df.iterrows():
    text += r"""
        \hline
            \multirow{1}{1cm}{\centering $INDEX!}& \multirow{1}{7.5cm}{\centering $NOME!} & \multirow{1}{7cm}{\centering \underline{\hspace{6cm}}} \\
        \hline
    """.replace('$INDEX!', str(i+1)).replace('$NOME!', row['name']).replace(r'_', r'\_').replace('$', r'\$')

text += '\n' + r'\end{longtable}' + '\n' + r'\end{document}'
```
